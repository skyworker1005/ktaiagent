# Azure AI Foundry 기초 

랭체인을 사용하지 않고 azure의 순수 라이브러리만으로 챗봇을 만들어봅시다.

## 1. Azure AI Foundry란?

**Azure AI Foundry**는 기업이 AI 애플리케이션을 쉽고 안전하게 만들 수 있도록 돕는 통합 플랫폼

### 필요성
- **통합 관리**: GPT-4 같은 모델뿐만 아니라, 검색(Search), 안전 필터(Safety), 평가(Evaluation) 도구를 한곳에서 쓸 수 있습니다.
- **SDK 통일**: `azure-ai-inference`라는 하나의 라이브러리로 다양한 모델(OpenAI, Llama, Phi 등)을 똑같은 방식으로 쓸 수 있습니다.

## 2. 환경 설정 (Environment Setup)

### 필요한 변수들
1.  `AZURE_INFERENCE_ENDPOINT`: 모델이 배포된 주소 (예: `https://kt-openai...`)
2.  `AZURE_INFERENCE_KEY`: 모델에 접근할 수 있는 비밀번호

이 값들은 Azure Portal 혹은 Azure AI Foundry에서 확인할 수 있습니다.

## 3. 코드 구성


```python
# %pip install azure-ai-inference
```

```python
import os
from dotenv import load_dotenv
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

load_dotenv("../../env")

endpoint = os.getenv("END_POINT", "").rstrip("/")
api_key  = os.getenv("AZURE_OPENAI_API_KEY")
model    = os.getenv("MODEL_NAME")

# Azure OpenAI: /openai/deployments/<model> 경로 자동 보정
if ".openai.azure.com" in endpoint and "/openai/deployments/" not in endpoint:
    endpoint = f"{endpoint}/openai/deployments/{model}"

client = ChatCompletionsClient(endpoint=endpoint, credential=AzureKeyCredential(api_key))
print(f"Endpoint: {endpoint}")
print(f"Model   : {model}")
```

### 클라이언트로 질문 보내기

이제 `client.complete` 함수를 사용해서 질문을 보냅니다.


```python
response = client.complete(
    messages=[
        SystemMessage(content="당신은 AI 기술을 아주 쉽게 설명해주는 친절한 선생님입니다."),
        UserMessage(content="'LLM(거대 언어 모델)'이 뭐야? 초등학생도 이해할 수 있게 설명해줘."),
    ]
)

print("답변:")
print(response.choices[0].message.content)
```
